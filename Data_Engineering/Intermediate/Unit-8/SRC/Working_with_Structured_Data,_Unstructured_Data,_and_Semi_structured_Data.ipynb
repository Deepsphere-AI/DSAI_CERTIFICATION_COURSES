# **Structured Data**

Structured data is defined as a form of data with a standard format. The Structured data has a very well-defined structure, order and consistency and can be easily read by software programs or functions. The Structured data is usually stored in databases tables.

# **Unstructured Data**

Unstructured data, as it suggests, is a data in the unstructured form, meaning it does not have any structure, order or consistency. In today's business world, typically, every organization generates data in unstructured form, be it text files, images, audio files or video files; the data generated is massive. 

## **Semi-Structured Data**

We can extract data from typically any data source, including database tables, flat files, applications, images, videos etc. To support these data structures, there are diverse data formats. JSON, XML etc., have become quite popular for transportation and storage of data. The Semi-structured datasets do not have any standard format. It does not follow the structure of a relational database model, also does not possess a standard schema

```python
from pdfquery import PDFQuery
import pdfminer
from pdfminer.pdfpage import PDFPage, PDFTextExtractionNotAllowed
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.layout import LAParams
from pdfminer.converter import PDFPageAggregator

import matplotlib.pyplot as plt
from matplotlib import patches
%matplotlib inline

import pandas as pd

def vAR_extract_page_layouts(vAR_file='/content/drive/MyDrive/Data Engineering/Intermediate/Unit-8/Source Data/example_anonymous.pdf'):
    """
    Extracts LTPage objects from a pdf file.
    modified from: http://www.degeneratestate.org/posts/2016/Jun/15/extracting-tabular-data-from-pdfs/
    Tests show that using PDFQuery to extract the document is ~ 5 times faster than pdfminer.
    """
    vAR_laparams = LAParams()
    
    with open(vAR_file, mode='rb') as vAR_pdf_file:
        print("Open document %s" % vAR_pdf_file.name)
        vAR_document = PDFQuery(vAR_pdf_file).doc

        if not vAR_document.is_extractable:
            raise PDFTextExtractionNotAllowed

        vAR_rsrcmgr = PDFResourceManager()
        vAR_device = PDFPageAggregator(vAR_rsrcmgr, laparams=vAR_laparams)
        vAR_interpreter = PDFPageInterpreter(vAR_rsrcmgr, vAR_device)

        vAR_layouts = []
        for vAR_page in PDFPage.create_pages(vAR_document):
            vAR_interpreter.process_page(vAR_page)
            vAR_layouts.append(vAR_device.get_result())
    
    return vAR_layouts

#example_file = "data/example_anonymous.pdf"
#page_layouts = extract_page_layouts(example_file)
#print("Number of pages: %d" % len(page_layouts))

extract_page_layouts(file='/content/drive/MyDrive/Data Engineering/Intermediate/Unit-8/Source Data/example_anonymous.pdf')
```
