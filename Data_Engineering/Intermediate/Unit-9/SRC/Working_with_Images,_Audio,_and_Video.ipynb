```python
class Working_with_Images:
  def __init__(self):
    return None
    
  def read_data_from_drive(self):
    drive.mount('/content/drive',force_remount=True)
    self.Image = '/content/drive/MyDrive/Data Engineering/Intermediate/Unit-9/Source Data/Apple.jpg'

  def read_image(self):    
    self.Image = cv2.imread(self.Image,1)
    cv2_imshow(self.Image)

  def Image_Characteristics(self):    
    print(self.Image.shape)
    
  def Rotate_Image(self):
    self.height, self.width = self.Image.shape[0:2]
    self.rotationMatrix = cv2.getRotationMatrix2D((self.width/2, self.height/2), 90, .5)
    self.rotatedImage = cv2.warpAffine(self.Image, self.rotationMatrix, (self.width, self.height))
    #print(self.rotatedImage)
    print(cv2_imshow(self.rotatedImage))

  def crop_image(self):
    self.startRow = int(self.height*.15)
    self.startCol = int(self.width*.15)
    self.endRow = int(self.height*.85)
    self.endCol = int(self.width*.85)
    self.croppedImage = self.Image[self.startRow:self.endRow, self.startCol:self.endCol]
    print(cv2_imshow(self.croppedImage))

  def Resize_image(self):
    self.newImg = cv2.resize(self.Image, (0,0), fx=0.75, fy=0.75)
    print(cv2_imshow(self.newImg))
    
  def run_all(self):
    self.read_data_from_drive()
    self.read_image()
    #self.Image_Characteristics()
    #self.Rotate_Image()
    #self.crop_image()
    #self.Resize_image()

vAR_obj = Working_with_Images()
vAR_obj.run_all()
```
