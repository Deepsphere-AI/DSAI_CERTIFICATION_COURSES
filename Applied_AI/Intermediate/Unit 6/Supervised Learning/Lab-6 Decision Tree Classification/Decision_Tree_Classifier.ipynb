```python
import pandas as vAR_pd
import numpy as vAR_np
import matplotlib.pyplot as vAR_plt
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
%matplotlib inline
from sklearn.tree import DecisionTreeClassifier  
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder


vAR_dataset = vAR_pd.read_excel("/content/Customer Churn.xlsx")

class DecisionTree:
  def dataassign(self,vAR_numcols,vAR_categcols):    
     print(vAR_numcols,vAR_categcols)
  def scaling(self,vAR_scale):
     print(vAR_scale)
  def encoding(self,vAR_encode):
     print(vAR_encode) 
  def modeltraining(self,vAR_itrain, vAR_itest, vAR_otrain, vAR_otest):
    vAR_model = DecisionTreeClassifier()
    vAR_model.fit(vAR_itrain, vAR_otrain)
    vAR_tpredict = vAR_model.predict(vAR_itrain)
    vAR_taccuracy = accuracy_score(vAR_otrain, vAR_tpredict)  
    vAR_testpredict = vAR_model.predict(vAR_itest)
    vAR_testaccuracy = accuracy_score(vAR_otest, vAR_testpredict)     
    print(vAR_taccuracy,vAR_testaccuracy)

def main():
    #values assigning
    vAR_input = vAR_dataset.drop(columns=["Churn","Churn Days","Churn Reasons","CustomerID","Supply Date"],axis=1)
    vAR_output = vAR_dataset["Churn"].values
    vAR_numcols = vAR_input.select_dtypes(include=vAR_np.number).columns
    vAR_categcols = vAR_input.select_dtypes('object').columns.tolist()
    #scaling
    vAR_scaler = StandardScaler()
    vAR_scaler.fit(vAR_input[vAR_numcols])
    vAR_input[vAR_numcols] = vAR_scaler.transform(vAR_input[vAR_numcols])
    vAR_scale = vAR_input[vAR_numcols]
    #encoding
    vAR_encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
    vAR_encoder.fit(vAR_input[vAR_categcols])
    vAR_encodecols = list(vAR_encoder.get_feature_names(vAR_categcols))
    vAR_input[vAR_encodecols] = vAR_encoder.transform(vAR_input[vAR_categcols])
    vAR_encoding = vAR_input[vAR_encodecols]
    #droping columns
    vAR_input = vAR_input.drop(columns='Product',axis=1)
    vAR_input = vAR_input.drop(columns='Service Start Date',axis=1)
    #training and test dataset
    vAR_itrain, vAR_itest, vAR_otrain, vAR_otest = train_test_split(vAR_input, vAR_output, test_size = 0.2,random_state=2)

    vAR_Obj = DecisionTree()

    print("the dataset numerical and object variables are is:")
    vAR_DA = vAR_Obj.dataassign(vAR_numcols,vAR_categcols)    
    vAR_DA
    
    print(" After scaling")
    vAR_sc = vAR_Obj.scaling(vAR_scale)
    vAR_sc

    print("After Encoding")
    vAR_ec = vAR_Obj.encoding(vAR_encoding)
    vAR_ec

    print("the model accuracy is")
    vAR_M = vAR_Obj.modeltraining(vAR_itrain, vAR_itest, vAR_otrain, vAR_otest)
    vAR_M






if __name__=="__main__":
    main()     
```
