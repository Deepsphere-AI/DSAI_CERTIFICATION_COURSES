```python
import pandas as vAR_pd
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures

vAR_dataset = vAR_pd.read_csv('/content/Boston.csv')

print(vAR_dataset)

class Ridgelasso:
  def datavisual(self,vAR_input,vAR_output):
    print(vAR_input,vAR_output)
  def Ridge(self,vAR_itrain,vAR_otrain,vAR_itest,vAR_otest):
    vAR_ridge = Ridge()
    vAR_ridge.fit(vAR_itrain,vAR_otrain)
    vAR_train = vAR_ridge.score(vAR_itrain,vAR_otrain)
    vAR_test = vAR_ridge.score(vAR_itest,vAR_otest)
    print(vAR_train,vAR_test)
  def Lasso(self,vAR_itrain,vAR_otrain,vAR_itest,vAR_otest):
    vAR_Lasso = Lasso(alpha=0.003)
    vAR_Lasso.fit(vAR_itrain,vAR_otrain)
    vAR_train = vAR_Lasso.score(vAR_itrain,vAR_otrain)
    vAR_test = vAR_Lasso.score(vAR_itest,vAR_otest)
    print(vAR_train,vAR_test)
    
def main():
    vAR_Obj = Ridgelasso()   
    #the dataset used here is a boston house price and we use it to predict the houses price
    vAR_input = vAR_dataset.drop(columns='medv')
    vAR_output = vAR_dataset["medv"] 

    vAR_input = preprocessing.scale(vAR_input)
    vAR_output = preprocessing.scale(vAR_output)
    vAR_itrain, vAR_itest, vAR_otrain, vAR_otest = train_test_split(vAR_input, vAR_output, test_size = 0.2,random_state=2)
    vAR_model1 = LinearRegression()
    vAR_model1.fit(vAR_itrain,vAR_otrain)
    vAR_modelp = PolynomialFeatures()    
    vAR_pinput = vAR_modelp.fit_transform(vAR_input)
    vAR_itrain1, vAR_itest1, vAR_otrain1, vAR_otest1 = train_test_split(vAR_pinput, vAR_output, test_size = 0.2,random_state=2)
    vAR_model1.fit(vAR_itrain,vAR_otrain)
    
    print("The input and output values are:")
    vAR_dv = vAR_Obj.datavisual(vAR_input,vAR_output)
    vAR_dv

    print("the score for the train and test  data using Ridge is:")
    vAR_r = vAR_Obj.Ridge(vAR_itrain,vAR_otrain,vAR_itest,vAR_otest)
    vAR_r
    
    print("the score for the train and test data using Lasso is:")
    vAR_l = vAR_Obj.Lasso(vAR_itrain,vAR_otrain,vAR_itest,vAR_otest)
    vAR_l

if __name__=="__main__":
    main()      
```
